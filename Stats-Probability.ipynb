```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

# How to calculate BINOMINAL DISTRIBUTION

```python
n=7
p=0.6
k = np.arange(0,8)
k
```

```python
binominal = stats.binom.pmf(k,n,p)
binominal
```

```python
plt.plot(k,binominal, 'o-')
plt.title('Binominal: n=%i p=%.2f' %(n,p), fontsize=20)
plt.xlabel('Number of accounts')
plt.ylabel('Probability of success')
plt.show()
```

# How to calculate POISSION DISTRIBUTION

```python
#average
rate = 3
n = np.arange(0,20)
n
```

```python
poission = stats.poisson.pmf(n,rate)
poission
```

```python
# what is the prob. of exactly 4 customers arrive in a minute
poission[4]
```

```python
# what is the prob. of more than 3 cutomers arrives in a minute
1 - (poission[0]+poission[1]+poission[2]+poission[3])
```

```python
plt.plot(n, poission ,'o-')
plt.title('Poission: $\lambda$= %i' % rate)
plt.xlabel('Number of customers arriving in a minute')
plt.ylabel('Probability of no. of cutomers arriving in a min')
plt.show()
```

# Normal distribution calculation

```python
# calculate z
z = (0.280 - 0.295)/0.025
z
```

```python
# cumulate distribution function gives value left of the curve or less than some value
stats.norm.cdf(z)
```

```python
# using original vlues
stats.norm.cdf(0.280, loc=0.295, scale=0.025)
```

```python
#calculate more than value i,e probability or area on right side of the curve is calculated ( 1- probability to left)
1 - stats.norm.cdf(0.350, loc=0.295, scale= 0.025)
```

```python
#In between two points
stats.norm.cdf(0.340, loc = 0.295, scale = 0.025) - stats.norm.cdf(0.260, loc = 0.295, scale = 0.025)
```

```python
#the probability of being between 1,2 and 3 of the mean
stats.norm.cdf(1) - stats.norm.cdf(-1)
```

```python
stats.norm.cdf(2) - stats.norm.cdf(-2)
```
