```python
#Visualisation 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(1)
random_values = np.random.randn(50)
sns.displot(random_values, kde=True, height=4, aspect=2)
plt.show()
```

```python
penguins_data = sns.load_dataset("penguins")
penguins_data.head()
```

```python
sns.displot(data=penguins_data, x="flipper_length_mm", kde=True, height=4, aspect=2) 
plt.show()
```

```python
sns.jointplot(data=penguins_data, x="bill_length_mm", y="bill_depth_mm", hue="species")
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(np.random.randn(20).reshape(4,5)) # random numbers of shape (4,5)
correlation_matrix = df.corr() # shows correlation
correlation_matrix
```

```python
ax_grid = sns.heatmap(correlation_matrix, cbar=0, linewidths=2,vmax=1, vmin=0, square=True, cmap='Blues', annot=True)
plt.show()
```

```python
sns.set_theme()  #Set aspects of the visual theme for given seaborn plot
uniform_values = np.random.rand(14, 16)
ax_grid = sns.heatmap(uniform_values, cbar=0, linewidths=2,vmax=1, vmin=0, square=True)
```

```python
sns.set_theme(style="darkgrid")
titanic_values = sns.load_dataset("titanic") # seaborn has inbuild varaity of datasets
ax_grid = sns.countplot(x="class", data=titanic_values)
```

```python
titanic_values.head()
```

```python
ax_grid = sns.countplot(x="class", hue="who", data=titanic_values) #to map plot aspects to different colors
```

```python
sns.pairplot(data=penguins_data, hue="species")
```

```python
#to make diagnolar histogram
sns.pairplot(data=penguins_data, hue="species", diag_kind="hist")
```

```python
tips_data = sns.load_dataset("tips")
tips_data.head()
```

```python
sns.boxplot(x=tips_data['total_bill'])
```

```python
ax_grid = sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips_data, palette="Set3")
```

```python
sns.scatterplot(x=tips_data['total_bill'], y=tips_data['tip'], hue=tips_data["time"])
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips_data,hue='smoker');
```

```python
ax_grid = sns.pointplot(x="time", y="total_bill", data=tips_data)
```

```python
sns.barplot(x="tip", y="day", data=tips_data)
```

```python
#to remove black error lines
sns.barplot(x="tip", y="day", data=tips_data,ci=None)
```
